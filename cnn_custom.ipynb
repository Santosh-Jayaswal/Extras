```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras import layers, models
```

```python
images_path = ["./images/top_left.png", "./images/top_right.png", "./images/bottom_left.png", "./images/bottom_right.png"]
images = [load_img(image) for image in images_path]
img_array = [img_to_array(image) for image in images]
data = np.stack(img_array, axis=0)
data /= 255

model = models.Sequential([
    layers.Conv2D(9, kernel_size=(3, 3), strides=(1, 1), activation='sigmoid', input_shape=(4, 4, 3)),
])


cnn_data = model.predict(data)

n_maps = cnn_data.shape[0]
fig, axes = plt.subplots(1, n_maps, figsize = (20, 5))

for i in range(n_maps):
    map_img = cnn_data[i, :, :, 0]

    axes[i].imshow(map_img)

# cnn_data
```
